# Model Selection and Cross Validation

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Purchased_Dataset.csv')
x = df[['Age', 'EstimatedSalary']]
y = df['Purchased']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 5)
knnClassifier = KNeighborsClassifier(n_neighbors = 5)
knnClassifier.fit(x_train, y_train)
y_pred = knnClassifier.predict(x_test)
metrics.accuracy_score(y_test, y_pred)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 7)
knnClassifier = KNeighborsClassifier(n_neighbors = 5)
knnClassifier.fit(x_train, y_train)
y_pred = knnClassifier.predict(x_test)
metrics.accuracy_score(y_test, y_pred)
```

## Cross Validation Score (K -Fold Technique)

> It helps us to calculate the accuracy score of trained model with multiple algorithms and sample data sets

* We can decide the sample subsets based on data sample.
* We can use multiple algorithms and check the best score



```python
from sklearn.model_selection import cross_val_score

knnClassifier = KNeighborsClassifier(n_neighbors = 4)

# cv (cross value) is number of sample data subsets 
print(f"All accuracy scores: {cross_val_score(knnClassifier, x, y, cv = 10, scoring = 'accuracy')}")

# Check the mean of all scores
print(f"Means= {cross_val_score(knnClassifier, x, y, cv = 10, scoring = 'accuracy').mean()}")
```

### Test the model performance/ accuracy score with other algorithms

* I have test the model performance score with logistic regression. it produce the accuracy score is 0.64 but KNN produce 0.79, So KNN is perform better than logistic regression. We can test with more algorithms.

```python
from sklearn.linear_model import LogisticRegression
lgreg = LogisticRegression()

print(cross_val_score(lgreg, x, y, cv = 10, scoring = 'accuracy').mean())
```
